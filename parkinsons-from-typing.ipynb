### Predicting the possibility that a person has Parkinson's based on the typing patterns observed. Original data set has been merged and processed into parkinsons_final.csv.


```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix,accuracy_score,mean_squared_error
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('../input/typing-parkinsons/parkinsons_final.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df_copy=df.copy()
```

```python
df_copy.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df_copy.shape
```

```python
df_copy.head()
```

```python
true_false={True:1,False:0}
df_copy['Tremors']=df['Tremors'].map(true_false)
df_copy['Parkinsons']=df['Parkinsons'].map(true_false)
df_copy['Levadopa']=df['Levadopa'].map(true_false)
df_copy['DA']=df['DA'].map(true_false)
```

```python
df_copy.head()
```

```python
df_copy.isnull().sum()
```

```python
df_copy['Age']=[(2021-i) for i in df_copy['BirthYear']]
```

```python
df_copy['Age']=df_copy['Age'].fillna('72')
```

```python
df_copy['Impact']=df_copy['Impact'].fillna('Medium')
```

```python
df_copy.drop('DiagnosisYear',axis=1,inplace=True)
```

```python
df_copy.isnull().sum()
```

```python
df_copy['Sided'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_copy['Gender'] = le.fit_transform(df_copy['Gender'])
df_copy['UPDRS'] = le.fit_transform(df_copy['UPDRS'])
df_copy['Sided'] = le.fit_transform(df_copy['Sided'])
df_copy['Impact'] = le.fit_transform(df_copy['Impact'])
```

```python
df_copy['MAOB'] = le.fit_transform(df_copy['MAOB'])
```

```python
df_copy['Age']=df_copy['Age'].astype('int')
```

```python
df_copy.info()
```

```python
X=df_copy.drop(['Parkinsons','BirthYear'],axis=1)
y=df_copy['Parkinsons']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=0)
```

## Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors=25)
model_knn.fit(X_train,y_train)
```

```python
result_knn=model_knn.predict(X_test)
mse_knn= mean_squared_error(y_test,result_knn,squared=False)
print("Mean Squared Error:",mse_knn)
cm=confusion_matrix(y_test, result_knn)
print("Confusion Matrix:\n",cm)
accuracy_score_knn= accuracy_score(y_test,result_knn)
print("Accuracy Score:",accuracy_score_knn)
```

## Using SVC

```python
from sklearn.svm import SVC
model_svc = SVC(kernel = 'rbf',gamma='auto')
model_svc.fit(X_train,y_train)
```

```python
result_svc=model_svc.predict(X_test)
mse_svc= mean_squared_error(y_test,result_svc,squared=False)
print("Mean Squared Error:",mse_svc)
cm=confusion_matrix(y_test, result_svc)
print("Confusion Matrix:\n",cm)
accuracy_score_svc= accuracy_score(y_test,result_svc)
print("Accuracy Score:",accuracy_score_svc)
```

## Using XGBoost

```python
from xgboost import XGBClassifier
model_XGB = XGBClassifier()
model_XGB.fit(X_train,y_train)
```

```python
result_xgb=model_XGB.predict(X_test)
mse_xgb= mean_squared_error(y_test,result_xgb,squared=False)
print("Mean Squared Error:",mse_xgb)
cm=confusion_matrix(y_test, result_xgb)
print("Confusion Matrix:\n",cm)
accuracy_score_xgb= accuracy_score(y_test,result_xgb)
print("Accuracy Score:",accuracy_score_xgb)
```

### Achieved highest accuracy using XGBoost amongst these models. Data-set quite short hence cannot implement deep learning models.
